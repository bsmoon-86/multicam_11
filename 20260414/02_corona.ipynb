```python
import pandas as pd 
```

```python
# csv 폴더 안에 있는 corona.csv 파일 로드 
# 현재 작업중인 폴더(./)(생략이 가능) -> 상위 폴더로 이동(../) 
# -> csv 하위 폴더로 이동(csv/) -> corona.csv
# 상대경로는 내 파일의 위치에서부터 시작
corona = pd.read_csv("../csv/corona.csv")
```

```python
# 데이터의 확인 : head(), tail()
corona.head()
```

```python
corona.tail()
```

```python
# 데이터프레임에서 컬럼들의 목록 확인 
# DataFrame라는 Class에는 3개 변수 -> values, index, columns
corona.columns
```

```python
# Unnamed: 0 컬럼은 인덱스로 사용이 됬었구나 -> 필요가 없다. -> 해당 컬럼의 데이터는 제거
# seq 컬럼의 데이터는 게시글의 번호 -> 제거 
# 특정 데이터를 제거(drop())
# dropna() -> 매개변수 축 설정(axis) -> 인덱스를 제거할것인가? 컬럼을 제거할것인가?
#           -> 매개변수 inplace -> 원본의 데이터를 변화 시킬것인가
#           -> 결측치를 제거한다
# drop() -> axis, inplace 매개변수 존재  
#       -> 첫번째 인자는 어떠한 것을 제거할것인가?
#       -> drop() 함수에서 기본 설정은 인덱스를 제거 
corona.drop(
    ['Unnamed: 0','seq'], axis=1, inplace=True
)
# 우리가 제거할 컬럼의 개수가 2개
# drop(inplace=True)의 함수를 실행하면 -> 첫번째 실행에서 정상 처리가 되더라도 
# -> 같은 블록을 다시 실행하면 에러 발생 ( 해당 데이터가 제거되고 다시 제거하려니 없다. )
```

```python
corona.head(1)
```

```python
# 컬럼의 이름을 변경 
# case1 : rename() 함수를 이용하여 컬럼의 이름을 변경 
# 특정 컬럼의 이름을 변경할때 사용
corona.rename(
    columns = {
        'createDt' : '등록일시', 
        'deathCnt' : '총사망자', 
        'decideCnt' : '총확진자', 
        'stateDt' : '기준일', 
        'stateTime' : '기준시간', 
        'updateDt' : '수정일시', 
        'accExamCnt' : '누적의심자', 
        'accDefRate' : '누적확진율'
    }
)
```

```python
# case2 : Class 내부의 columns 변수에 접근해서 다른 데이터를 대입 
# ( 데이터를 변경하기 전에 확인은 불가능 ) 
# -> 이 방법은 모든 컬럼의 이름을 변경할때 주로 사용
# 주의사항 : 컬럼의 개수가 같은 자료형 데이터를 대입 (길이가 다르면 에러 발생)
corona.columns = ['등록일시', '총사망자', '총확진자', '기준일', 
                  '기준시간', '수정일시', '누적의심자', '누적확진율']
```
