```python
# 조건문(if문)
# if 뒤의 조건식에 따라 참인 경우에 실행하는 코드와 
# 거짓인 경우 실행하는 코드를 나눠서 작성 
# python -> if 조건식 :
# 일반적인 언어 -> if (조건식) {  
a = 11
# if문의 시작 
if a < 10:
    # 조건식이 참인 경우 실행할 코드를 작성 
    print('a는 10보다 작다')
# 거짓인 경우 실행할 코드 
else:
    print('a는 10보다 크거나 같다')
# if ~ else 문에서는 하나의 지역의 코드만 실행이 된다. 
```

```python
# 조건문에서 조건식이 여러개로 나눠지는 경우 
score = 73  # C 학점

# 1번 조건식 : score가 90점 이상이라면 
if score >= 90:
    # score가 90점 이상인 경우에만 실행
    print('A학점')
else:
    # score가 90점 미만인 경우 실행이되는 지역 
    if score >= 80:
        # score가 90점 미만이고 80점 이상인 경우 실행
        print('B학점')
    else:
        # score가 80점 미만인 경우
        print('C학점')
```

```python
# if ~ else를 사용 -> if 뒤의 조건식의 개수를 변경 
if score >= 90:
    print('A학점')
# B학점의 조건식 : 
# score가 90점 미만(조건식1) 그리고(and) 80점 이상(조건식2)
if (score < 90) and (score >= 80):
    print('B학점')
# C학점의 조건식 :
# score가 80점 미만 그리고 70점 이상
if (score < 80) and (score >= 70):
    print('C학점')
# F학점의 조건식 : 
# score가 70점 미만
if score < 70:
    print('F학점')
```

```python
# if ~ elif ~ else문
# input() -> 코드를 실행할때 입력값을 받아온다.( 문자형으로 데이터를 반환 )
score = int(input('점수를 입력하시오 : '))

if score >= 90:
    print('A학점')
elif score >= 80:
    print('B학점')
elif score >= 70:
    print('C학점')
else:
    print('F학점')

print('프로그램 종료')
```

```python
# while문 

# 시작값 지정 
i = 0
# while 조건식 : 조건식이 거짓이 될때까지 반복 실행
# 조건식이 거짓이 될수 있도록 반복 실행 코드에 시작값을 증가/감소 
# break문을 이용해서 반복문을 강제 종료
# 시작값으로 조건식을 사용했을때 조건식은 참
while i < 10:
    # 초기값 i를 1씩 증가 
    i = i + 1
    print(i)
```

```python
result = 0
i = 1

# while i < 11:
while i <= 10:
    # 첫번째 반복 : result = 0 / i = 1 -> result = 0 + 1 -> i = 1 + 1 
    #               // result = 1, i = 2
    # 두번째 반복 : result = 1 / i = 2 -> result = 1 + 2 -> i = 2 + 1
    #               // result = 3, i = 3
    result = result + i
    i = i + 1
print(result)
```

```python
result = 0
i = 1

# while i < 11:
while i <= 10:
    # 첫번째 반복 : result = 0 / i = 1 -> i = 1 + 1 -> result = 0 + 2
    #               // result = 2, i = 2
    # 두번째 반복 : result = 2 / i = 2 -> i = 2 + 1 -> result = 2 + 3
    #               // result = 5, i = 3
    # 열번째 반복 : result = 54 / i = 10 -> i = 10 + 1 -> result = 54 + 11
    #               // result = 65, i = 10
    i = i + 1
    result = result + i
print(result)

# 일반적으로 while문에서는 초기값의 증가 식을 마지막에 작성 ( 까먹고 안 적어서 무한루프 )
```
