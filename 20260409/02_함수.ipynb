```python
# sort() -> 오름차순 / 내림차순 정렬 을 할수 있는 새로운 함수 
# 오름차순 정렬 -> min()
# 내림차순 정렬 -> max()

# 함수 선언 
    # 매개변수 2개 
        # values 인자의 개수가 가변인 형태
        # reverse 기본값이 False(오름차순 정렬) // True로 변경한다면 (내림차순 정렬)

def custom_sort( *values, reverse = False ):
    # 정렬이 되는 데이터들을 저장할 공간(빈 리스트 생성)
    result = []
    # values의 데이터가 수정을 할수 있도록 list로 변환 
    values = list(values)
    # values의 길이만큼 반복하는 반복문 생성 : values의 type이 tuple이기때문에
    for i in range(len(values)):
        # reverse의 값이 True라면  max() 함수를 이용해서 values의 최대값을 추출
        if reverse:
            value = max(values)
        else:
            value = min(values)
        result.append(value)
        values.remove(value)
    return result
```

```python
custom_sort(3,2,4,1, reverse = True)
```
