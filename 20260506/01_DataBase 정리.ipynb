### sql 총 정리 
- DDL 
    - CREATE
        - 데이터베이스나 테이블을 생성하는 명령어 
            - `CREATE DATABASE name`
            - `CREATE TABLE table (COL1 TYPE OPTION, ...)`
        - 테이블 구성에서 Datatype
            - 문자형
                - char(bytes)
                    - 고정형 데이터를 담는 공간 
                    - 데이터의 길이에 상관없이 특정 공간을 사용
                - varchar(bytes)
                    - 유동적인 길이의 데이터를 담는 공간 
                    - 데이터의 길이에 +1byte 공간이 생성
                - text
                    - 길이가 매우 긴 데이터를 담는 공간 
            - 숫자형 
                - int
                    - 숫자를 담는 공간
                    - 매우 작은 범위의 숫자는 tinyint 사용
        - Option
            - 제약 조건 
                - not null
                    - 결측치를 허용하지 않는다. 
                - unique
                    - 같은 필드내에서 중복 값을 허용하지 않는다. 
                - default
                    - 값이 들어오지 않을때 기본값을 설정 
                - primary key
                    - not null과 unique 조건을 모두 포함하는 기본키( 테이블에서 하나의 필드는 기본키를 가져야한다. ) 
                - auto_increment
                    - 자동 증가
                    - 해당 필드는 데이터를 대입하지 않고 자동적으로 1씩 증가하여 데이터가 대입
    - DROP
        - 데이터베이스나 테이블을 제거하는 명령어
            - `DROP DATABASE name`
            - `DROP TABLE table`
    - ALTER
        - 기존에 생성되어 있는 데이터베이스나 테이블의 구조를 변경하는 명령어 
            - `ALTER DATABASE name OPTION`
            - `ALTER TABLE table OPTION`
                - ADD 
                    - 필드를 추가하는 OPTION
                - DROP 
                    - 필드를 제거하는 OPTION
                - MODIFY COLUMN 
                    - 필드를 수정하는 OPTION
- DML (CRUD)
    - INSERT
        - 생성이 된 테이블에 데이터를 대입하는 명령어 
            - `INSERT INTO table VALUES (data1, data2, ...)
                - 테이블의 모든 필드에 데이터를 대입 
            - `INSERT INTO table(COL1, COL2, ...) VALUE (data1, data2, ...)
                - 테이블에서 특정 필드에 데이터를 대입 
    - UPDATE
        - 대입이 된 데이터를 수정하는 명령어 
            - `UPDATE table SET COL1 = data1`
                - 특정 필드의 데이터들은 모두 수정 
            - `UPDATE table SET COL1 = data1 WHERE 조건식`
                - 특정 필드의 조건식에 맞는 데이터를 찾아서 수정
    - DELETE 
        - 테이블 안에 데이터를 제거 하는 명령어
            - `DELETE FROM table`
                - 해당 테이블의 모든 데이터를 제거 
            - `DELETE FROM table WHERE 조건식`
                - 특정한 조건식에 맞는 데이터를 제거 
        - DROP 명령어와 DELETE 명렁어의 차이는? 
            - 테이블의 존재 유무  
                - DELETE는 테이블 안에 데이터를 제거 
                - DROP은 테이블을 제거 (테이블이 남아있지 않음)
    - SELECT 
        - 데이터를 조회하는 명령어 ( 데이터의 변화가 존재하지 않는다. )
            - `SELECT * FROM table`
                - 테이블의 모든 행과 모든 열을 조회
            - `SELECT COL1, COL2 FROM table`
                - 테이블에서 특정 열을 선택하여 조회
            - `SELECT * FROM table WHERE 조건식`
                - 테이블에서 특정 조건식에 맞는 행 데이터들을 조회
    - 조건식
        - 일반적인 비교연산자를 이용해서 조건식을 생성 (=, !=, >, <, >=, <=)
        - IN, NOT IN
            - 포함 여부
        - BETWEEN 
            - 사이값
        - LIKE
            - ~같은 -> 특정 문자가 포함되어있는가?
                - S% -> S로 시작하는 조건식
                - %S% - S가 포함되는 조건식
                - %S -> S로 끝이 나는 조건식
    - ORDER BY
        - SELECT문에서 정렬을 변경하는 OPTION
            - ASC -> 오름차순 정렬 
            - DESC -> 내림차순 정렬
    - UNION
        - 두 개의 조회된 테이블을 단순한 행 결합 OPTION
        - `SELECT * FROM table UNION SELECT * FROM table2`
    - JOIN 
        - 두 개의 테이블을 특정 조건에 맞춰서 열 결합 OPTION
        - 종류 : MYSQL은 LEFT, RIGHT, INNER
        - `SELECT * FROM table INNER JOIN table2 ON table.COL = table2.COL`
    - GROUP BY
        - 조회된 테이블에서 그룹화( 값은 값끼리 묶여서 표현 )
        - `SELECT SUM(COL1) FROM table GROUP BY COL2`
        - HAVING OPTION
            - 그룹화된 데이터에서 행을 필터 
            - WHERE과의 차이는?
                - WHERE은 그룹화 하기 전에 행 필터, HAVING은 그룹화 후 행 필터
        - 실행 순서 
            - FROM -> WHERE -> GROUP BY -> HAVING -> SELECT -> DISTINCT -> ORDER BY -> LIMIT
    - 별칭 설정 
        - COL|TABLE as 별칭
        - COL|TABLE 별칭
        - 별칭을 생성하면 별칭으로 명령어 안에서는 사용
- DCL (데이터베이스의 보안 관련)
    - GRANT
        - 권환 부여 
    - REVOKE
        - 권환 회수
