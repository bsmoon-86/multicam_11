1. 컬럼의 이름들을 변경 -> [국가, 맥주소비량, 위스키소비량, 와인소비량, 순수알콜, 대륙]
2. 파생변수 -> '총술소비량' -> 맥주소비량 + 위스키소비량 + 와인소비량
3. 파생변수 -> '도수' -> (순수알콜 / 총술소비량) * 100
4. 대륙 별 그룹화를 하여 도수의 평균을 구한다 -> 가장 독한 술을 먹는 대륙은 어디인가?

```python
# 컬럼 이름 변경 -> columns에 접근해서 데이터 대입 // rename() 함수를 이용해서 변경
df.columns = ['국가', '맥주소비량', '위스키소비량', '와인소비량', '순수알콜', '대륙']
```

```python
df.head(1)
```

```python
# 파생변수  -> dict에서 key:value 추가하는 방식과 같다. 
df['총술소비량'] = df['맥주소비량'] + df['위스키소비량'] + df['와인소비량']
```

```python
# sum() 함수를 활용
df['총술소비량2'] = df.loc[:, '맥주소비량' : '와인소비량'].sum(axis=1)
```

```python
df.head()
```

```python
# 총술소비량을 생성한 이유는 도수 컬럼을 생성하기 위함
df['도수'] = ( df['순수알콜'] / df['총술소비량'] ) * 100
```

```python
df.head()
```

```python
# 도수 결측치 부분은 0으로 채워준다. 
df['도수'] = df['도수'].fillna(0)
```

```python
# 도수가 0인 데이터를 제외하고 대륙 별 도수의 평균을 구해보자 
df.loc[
    df['도수'] > 0, 
    ['대륙', '도수']
].groupby('대륙').mean()
```

```python
# 과연 우리나라는 총 술소비량이 전세계 몇위쯤 될것인가?
# 총술소비량을 기준으로 내림차순 정렬 
# sort_values() 함수를 이용 -> 기본값으로는 오름차순 정렬 -> ascending 매개변수를 False 변경 
df.sort_values('총술소비량', ascending=False, inplace=True)
```

```python
# 순위 -> 위치 값에 + 1
# 국가의 이름이 'South Korea' 인 데이터를 추출해서 인덱스를 확인 
flag = df['국가'] == 'South Korea'
korea_index = df.loc[flag, ].index[0]
korea_rank = korea_index + 1
```

```python
# 한국의 총술소비량 데이터 저장 
korea_value = df.loc[korea_index, '총술소비량']
```

```python
# 모든 국가의 총술소비량 데이터를 이용해서 막대그래프 생성 
x = df.index
y = df['총술소비량']

flag = df['국가'] == 'South Korea'
korea_index = df.loc[flag, ].index[0]
korea_rank = korea_index + 1
korea_value = df.loc[korea_index, '총술소비량']

# 그래프 사이즈를 변경 
plt.figure(figsize=(20, 8))
# 국가별 막대 그래프를 그리고 한국의 막대의 색상을 변경한 뒤 주석을 달아서 한국이 몇위이고 수치가 몇인가?
bar_graph = plt.bar(x, y)

# 한국 위치의 막대를 선택해서 색상을 빨간색으로 변경 
bar_graph[korea_index].set_color('red')

# 그래프에 주석을 추가 (화살표로 주석이 나타내는 부분을 표시)
plt.annotate(
    f"South Korea \n rank : {korea_rank} \n 총술소비량 : {korea_value}",    # 주석에 들어갈 텍스트 지정
    xy = (korea_index, korea_value),                                      # 주석이 어느 부분을 설명하는가?(좌표)
    xytext= ( korea_index + 10, korea_value + 150 ),                      # 주석 문구의 위치
    arrowprops= {                                                         # 주석과 해당 데이터의 위치를 화살표 연결
        'facecolor' : 'r',          # 화살표의 안쪽의 색상을 지정
        'edgecolor' : 'black',      # 화살표 외부의 색상을 선택
        'headwidth' : 20,           # 화살표 머리의 너비를 지정
        'headlength' : 20,          # 화살표 머리의 길이
        'width' : 10,               # 화살표 꼬리의 너비
        'shrink' : 0.1              # 화살표와 xy, xytext간의 간격 조정
    }
)

# plt.xticks(rotation = 45)
plt.show()
```

```python
korea_rank = korea_index + 1
```

```python
# 랭크 함수 -> rank() -> 오름차순정렬순으로 ranking 생성 
# ascending 매개변수를 False 변경하면 내림차순 정렬순으로 ranking 생성
# method 매개변수 -> 동점인 데이터들의 순위를 어떻게 매길것인가?
    # average(기본값) : 순위의 평균값을 생성 (2위 데이터가 2개라면 2.5위 )
    # min : 동점자 들의 낮은 순위로 생성 (스포츠에서 순위를 매기는 방식) ( 공동 2위, 다음 순위 4위 )
    # max : 돔점자 들의 높은 순위로 생성 ( 1위 -> 공동 3위 -> 4위 )
    # first : 데이터 상의 먼저 나온 순서로 순위를 지정 ( sort_values() -> reset_index() 방식과 동일 )
    # dense : min과 흡사, 동점자들은 낮은 순위로 지정, 다음 순위가 +1 순위 (공동 2위, 다음 순위가 3위)
```
