```python
import pandas as pd 
```

```python
# csv 폴더 안에 tran_1.csv, tran_2.csv 파일을 로드 
tran_1 = pd.read_csv("../csv/tran_1.csv")
tran_2 = pd.read_csv("../csv/tran_2.csv")
```

```python
# tran_1, tran_2 데이터를 결합 
    # 두개의 데이터프레임은 테이블의 구조가 같은 형태 
    # 단순한 행 결합 
    # 두개의 데이터프레임이 행 결합을 하면 rows의 개수는? -> 5000 + 1786 = 6786
# concat() 함수 이용
# pandas 안에 concat 함수 호출 --> 원본의 데이터라는 개념이 존재 x -> 
# 결과를 저장하려면 새로운 변수에 대입 
tran = pd.concat(
    [ tran_1, tran_2 ], axis = 0, ignore_index= True
)
```

```python
tran.info()
```

```python
# tran_d_1.csv, tran_d_2.csv 파일을 로드 
tran_d_1 = pd.read_csv("../csv/tran_d_1.csv")
tran_d_2 = pd.read_csv("../csv/tran_d_2.csv")
```

```python
# tran_d_1, tran_d_2 데이터프레임을 단순 행 결합 
tran_d = pd.concat(
    [tran_d_1, tran_d_2], axis=0, ignore_index=True
)
```

```python
# tran과 tran_d 데이터프레임을 결합 
# 테이블의 구조는 다르게 구성 -> 단순한 행결합 불가능 
# 인덱스의 개수가 다르다 -> 6786 // 7144개의 데이터 
# 두개의 데이터프레임이 공통적으로 가지고 있는 데이터 컬럼은 존재하는가?
    # transaction_id 컬럼은 공통적으로 가지고 있다 -> 조인결합이 가능
# tran의 인덱스의 수는 6786개
# tran_d의 인덱스의 수는 7144개 
    # 두개의 데이터프레임을 조인 결합 했을때 -> 인덱스의 수는 어떻게 나올것인가?
    # 6786 // 7144 --> how 에 따라 다르게 나올것인가?
pd.merge(
    tran, tran_d, 
    # 조건 ( 두개의 데이터프레임이 공통적으로 가지고 있는 컬럼의 이름 )
    on = 'transaction_id',
    # 결합의 방식은 어떻게 할것인가?
    how = 'outer'
)
```
