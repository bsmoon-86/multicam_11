## matplotlib
- pyplot(plt)를 이용하여 도화지를 생성 
- plt 안에 함수들을 이용하여 그래프들을 생성 
- plt -> 그래프가 그려지는 캔버스(도화지) , 캔버스 안에 배경이 투명한 레이어들을 이용하여 그래프를 완성 
- 그래프를 그리고(plot, bar, scatter, ...), option들을 이용하여 그래프의 모양을 변경
- 범례 추가 : legend() 
- 격자선 추가 : grid() 
- 주석 추가 : annotate()
- 캔버스 설정 : figure(figsize = (20, 8)) 
- 한글 깨짐 방지 
    - 기본적인 matplot에서는 한글이 지원하지 않은 폰트를 사용
    - plt.rc('font', family='Malgun Gothic') : 윈도우
    - plt.rc('font', family='AppleGothic') : MAC
- 마이너스 기호 깨짐 방지
    - plt.rc('axes', unicode_minus=False)
- 캔버스에서 영역을 나눠주는 함수 
    - subplot(행의개수, 열의개수, 위치)
        - 이미 생성이 된 캔버스에서 영역을 나눠준다. 
        - 특정 영역을 선택하고 그래프를 그려준다.
        - ex) plt.subplot(2, 1, 1) -> plt.plot()
    - subplots(행의 개수, 열의 개수)
        - subplot()과의 차이는 subplots()는 영역을 나눈 캔버스를 따로 생성 ( 생성이된 캔버스를 변수에 저장 )
        - ex) fig, axes = plt.subplots(2, 1) -> axes[1][0].plot()
- show()
    - 그래프를 출력하는 함수 
    - 그래프를 출력하고 캔버스를 초기화
- savefig()
    - 그래프를 그림으로 저장하는 함수 
    - show() 함수가 호출되기 전에 사용 (show() 함수 이후에 savefig() 함수를 사용하면 빈 캔버스 저장)
## seaborn
- matplotlib 라이브러리의 보조 라이브러리 
- 그래프를 시각화 하는데 데이터 분석에서 자주 사용이되는 라이브러리 
- 여러개의 그래프를 동시에 그려주거나 데이터간의 관계, 분포를 간단하게 그려줄수 있는 라이브러리 
- 그래프를 테스트하기 위해 샘플 데이터셋을 제공 
- 함수 목록 
    - 데이터의 분포를 확인하는 함수 : displot()
    - 산점도와 분포를 동시에 확인하는 함수 : jointplot()
    - 산점도 그래프와 회귀선을 추가 : lmplot()
    - 박스 플롯(이상치 체크) : boxplot()
    - 데이터의 밀집도와 박스를 확인(바이올린 플롯) : violinplot()
    - 여러 필드의 데이터들의 산점도와 히스토그램과 같은 그래프를 확인 : pairplot()
