### 데이터의 분할
- 분석 모델을 학습하고 성과를 확인하기 위해서 데이터를 Train, Test 세트로 나눠주는 작업 
- 독립 변수(X) -> 문제 , 종속 변수(y) -> 정답
- 독립 변수와 종속 변수 모두 Train, Test 데이터로 나눠준다. 
- 분석 방법이나 데이터의 양에 따라서 Train, Validation, Test로 나눠주는 경우도 발생 
- 데이터를 분할하는 일반적인 비율 
    - Train : Test = 7 : 3 // 8 : 2
        - 가장 일반적인 분할 비율
    - Train : Validation : Test = 6 : 2 : 2
        - 데이터의 양이 많은 경우 검증 셋을 생성하여 2번에 검증 작업 
    - kfold를 이용하면 여러번의 작업을 하여 검증이 가능
    - 유의할 점 
        - 데이터의 불균형 문제.
        - 데이터의 균형을 알맞게 분할하는 작업이 필요
- sklearn안에 train_test_split() 함수를 이용하여 데이터를 분할
    - train_test_split( X, y, test_size = None, random_state = None, shuffle = True, stratify = None )
        - X : 독립 변수( 문제 )
        - y : 종속 변수( 정답 )
        - test_size : 테스트데이터의 비율 (0부터 1사이의 값)
        - random_state : 임의의 데이터를 추출하는 과정에서 seed 값을 지정 (같은 값을 지정하면 규칙이 생성되어 같은 데이터를 확인)
        - shuffle : 데이터를 섞을것인가 ( 시계열 데이터셋인 경우에는 False )
        - stratify : 특정 변수를 지정하면 지정한 변수를 기준으로 하여 계층화 되어 해당 변수의 비율을 유지하도록 데이터 분할

### 데이터 스케일링
- 대부분 분석 알고리즘은 컬럼 간 데이터의 범위가 크게 차이나는 경우에는 정상적으로 작동 하지 않는다. ( 성능이 떨어진다. )
- 값의 범위가 작은 컬럼에 비해서 범위가 큰 컬럼이 종속 변수를 예측하는데 큰 영향을 준다고 판단 
- 스케일링 작업은 모든 컬럼의 값의 범위를 같게 만들어주는 작업 
- 스케일링 작업 전에 이상치 데이터를 대체하거나 삭제
- 스케일링 작업 순서 
    - train, test로 나눠져있는 경우  train에서 사용한 scaler를 test에서 사용(일반적인 방법)
    1. Scaler 선택
        - 해당 Scaler를 import 
    2. Scaler 객체(class) 생성
    3. train 데이터의 분포(범위)를 저장 ( fit() )
    4. train 데이터를 스케일링 ( transform() )
    5. test 데이터를 스케일링 ( transform() )
    6. 원본 데이터로 변환 ( inverse_transform() )

#### Standard Scaler
- 표준화 방식으로 가장 기본적인 스케일러 
- 평균이 0, 분산이 1인 정규분포로 스케일링 
- 최솟값과 최대값의 크기를 따로 제한하지 않아 이상치에 민감하기 때문에 이상치에 대한 확인 및 정제를 한 뒤 사용
- 일반적으로 회귀분석보다는 분류분석에서 사용

```python
from sklearn.preprocessing import StandardScaler
```

```python
# class 생성 
stdscaler = StandardScaler()
```

```python
# X_train3 데이터의 범위를 등록 
# Scaler에서 fit() -> 범위를 지정 // transform() -> 지정된 범위를 이용하여 스케일링
# fit_transform() -> fit() + transform()
# train data : fit() -> transform() || fit_transform()
# test data : transform()
stdscaler.fit(X_train3)
```

```python
# train 데이터 스케일링 (변환) -> 따로 제공(함수의 리턴 데이터를 이용하여 원본 데이터를 변화시키지 않고 유지)
X_train_sc = stdscaler.transform(X_train3)
```

```python
# test 데이터 스케일링 
X_test_sc = stdscaler.transform(X_test3)
```

```python
# 데이터들의 최솟값과 최댓값 평균, 표준편차를 출력하는 함수를 생성
def scaler_print(train, test):
    print(f"""
        Train Data : 
            Min : {round(train.min(), 2)}
            Max : {round(train.max(), 2)}
            Mean : {round(train.mean(), 2)}
            Std : {round(train.std(), 2)}
    """)
    print(f"""
        test Data : 
            Min : {round(test.min(), 2)}
            Max : {round(test.max(), 2)}
            Mean : {round(test.mean(), 2)}
            Std : {round(test.std(), 2)}
    """)
```

```python
scaler_print(X_train_sc, X_test_sc)
```

```python
x_t_s = stdscaler.fit_transform(X_train3)
x_te_s = stdscaler.fit_transform(X_test3)
```

```python
scaler_print(x_t_s, x_te_s)
```

```python
scaler_print(X_train3.values, X_test3.values)
```

#### Min-Max Scaler 
- 정규화 방식으로 컬럼의 데이터들을 0과 1 사이의 값으로 스케일링 하는 방식
- 최솟값 0 최대값 1
- 이상치 매우 민감함으로 이상치 대한 확인 및 정제 필수 
- 일반적으로 분류 분석보다는 회귀 분석에서 사용

```python
from sklearn.preprocessing import MinMaxScaler

mnscaler = MinMaxScaler()

X_train_sc = mnscaler.fit_transform(X_train3)
X_test_sc = mnscaler.transform(X_test3)

scaler_print(X_train_sc, X_test_sc)
```

#### Max Abs Scaler
- 최대절댓값과 0이 각각 1, 0이 되도록 스케일링을 하는 정규화 방식으로 모든 값은 -1부터 1 사이로 표현
- 스케일링 대상의 데이터가 모두 양수라면 MinMaxScaler와 동일
- 이상치 매우 민감하기 때문에 확인 및 정제 필요
- 일반적으로는 분류 분석보다는 회귀 분석에서 사용

```python
from sklearn.preprocessing import MaxAbsScaler
mascaler = MaxAbsScaler()
X_train_sc = mascaler.fit_transform(X_train3)
X_test_sc = mascaler.transform(X_test3)

scaler_print(X_train_sc, X_test_sc)
```

#### Robust Scaler
- 평균과 분산을 이용한 Standard 대신에 중앙값과 사분위수를 활용하는 방식
- 중앙값을 0으로 설정, IQR을 사용하여 이상치의 영향을 최소화
- quantile_range 매개변수( 기본값 [25, 75] )의 값을 조정하여 더 넓거나 좁은 범위의 값을 이상치로 판단하게 하여 정제할 수 있다. 

```python
from sklearn.preprocessing import RobustScaler
ruscaler = RobustScaler()
ruscaler2 = RobustScaler(quantile_range=(20, 80))
```

```python
ruscaler.fit(X_train3)
```

```python
ruscaler2.fit(X_train3)
```

```python
X_train_sc = ruscaler.transform(X_train3)
X_test_sc = ruscaler.transform(X_test3)

scaler_print(X_train_sc, X_test_sc)
```

```python
X_train_sc = ruscaler2.transform(X_train3)
X_test_sc = ruscaler2.transform(X_test3)

scaler_print(X_train_sc, X_test_sc)
```
