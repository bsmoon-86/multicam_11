### 데이터의 분할
- 분석 모델을 학습하고 성과를 확인하기 위해서 데이터를 Train, Test 세트로 나눠주는 작업 
- 독립 변수(X) -> 문제 , 종속 변수(y) -> 정답
- 독립 변수와 종속 변수 모두 Train, Test 데이터로 나눠준다. 
- 분석 방법이나 데이터의 양에 따라서 Train, Validation, Test로 나눠주는 경우도 발생 
- 데이터를 분할하는 일반적인 비율 
    - Train : Test = 7 : 3 // 8 : 2
        - 가장 일반적인 분할 비율
    - Train : Validation : Test = 6 : 2 : 2
        - 데이터의 양이 많은 경우 검증 셋을 생성하여 2번에 검증 작업 
    - kfold를 이용하면 여러번의 작업을 하여 검증이 가능
    - 유의할 점 
        - 데이터의 불균형 문제.
        - 데이터의 균형을 알맞게 분할하는 작업이 필요
- sklearn안에 train_test_split() 함수를 이용하여 데이터를 분할
    - train_test_split( X, y, test_size = None, random_state = None, shuffle = True, stratify = None )
        - X : 독립 변수( 문제 )
        - y : 종속 변수( 정답 )
        - test_size : 테스트데이터의 비율 (0부터 1사이의 값)
        - random_state : 임의의 데이터를 추출하는 과정에서 seed 값을 지정 (같은 값을 지정하면 규칙이 생성되어 같은 데이터를 확인)
        - shuffle : 데이터를 섞을것인가 ( 시계열 데이터셋인 경우에는 False )
        - stratify : 특정 변수를 지정하면 지정한 변수를 기준으로 하여 계층화 되어 해당 변수의 비율을 유지하도록 데이터 분할

```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# 샘플 데이터셋 로드 
iris_data = load_iris()
```

```python
iris_data['frame']
```

```python
# 데이터프레임으로 변경 
iris = pd.DataFrame(iris_data['data'], columns = iris_data['feature_names'])

# target 데이터 데이터프레임에 추가 
iris['class'] = iris_data['target']

iris.head()
```

```python
# target 데이터들을 변환 
iris['class'] = iris['class'].map(
    lambda x : iris_data['target_names'][x]
)
```

```python
iris.head()
```

```python
# class 컬럼의 데이터의 빈도수를 확인 
iris['class'].value_counts()
```

```python
# 종속 변수, 독립 변수 데이터를 나눠서 train_test_splot() 입력 
x = iris.drop('class', axis=1)
x
```

```python
y = iris['class']
y
```

```python
# train_test_split() --> 결과물( train_x, test_x, train_y, test_y )
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size= 0.3, random_state=42
)
```

```python
# 데이터의 개수를 확인 
print(f"X_trian : {X_train.shape}, X_test : {X_test.shape}")
print(f"y_train : {y_train.shape}, y_test : {y_test.shape}")
```
