## openapi 
- api
    - 웹 주소 
    - 특정 주소에 요청을 보내면 데이터를 돌려주는 주소 
        - web service -> 보내주는 데이터가 화면으로 구성되는 html 문서 
        - data api -> 보내주는 데이터가 json(python에서 dict 형태와 유사) // xml(tag 형태로 존재하는 데이터)
- 사용이 되는 라이브러리 
    - requests
        - 라이브러리 안에 get() 함수를 이용하여 서버에게 요청을 보낸다. 
        - 첫번째 인자
            - 요청을 보내는 주소 
        - params 매개변수 
            - 요청을 보낼때 주소 외의 데이터를 보내는 매개변수 
            - dict 형태로 데이터를 보낸다 
                - 해당 주소에서 원하는 데이터를 key : value 형태로 구성해서 데이터를 생성 
        - 요청이 정상적으로 작동을 하면 code 200을 확인 할 수 있다. 
- code 안에 보안상 노출이 되면 위험한 데이터들은 숨겨놓는다. 
    - dotenv 라이브러리를 이용하여 환경변수에 데이터를 등록한다. 
    - .env 파일을 생성 민감한 정보들을 입력 
    - load_dorenv() 함수를 이용하여 .env 파일을 환경변수에 등록 
    - os 라이브러리를 이용하여 환경변수를 가져온다 (os.get_env(변수명))
    - .env 파일은 github과 같은 공유 공간에는 업로드 하지 않는다.(.gitignore 파일로 업로드 방지)

## 크롤링
- 웹 데이터를 실시간으로 가져오기 위한 방법 
- 웹 주소마다 크롤링을 허용하는 주소가 따로 존재 (url + /robots.txt 확인이 가능)
- 라이브러리 
    - requests
        - 웹 서버에서 html 문서를 받아오는 기능
    - bs4
        - BeautifulSoup class를 이용
        - requests를 이용하여 html문서를 받고 해당 문서를 Tag를 기준으로 데이터에 접근하기 위한 Class
        - parsing 작업을 이용하여 데이터의 타입을 변환하고 Class 안에 함수들을 이용하여 데이터에 접근
            - soup.태그명
                - 문서 안에서 해당 태그의 첫번째 태그의 정보를 불러온다. 
            - soup.태그명.string
                - 첫번째 태그에서 콘텐츠(태그 안에 데이터)를 추출
                - `<p>Test</p>` --> Test 을 추출
            - soup.태그명[속성명]
                - 첫번째 태그에서 특정 속성의 값을 추출
                - `<a href="https://www.naver.com">네이버</a>` --> `https://www.naver.com` 을 추출
            - 내장 함수 목록 
                - find()
                    - 문서 안에서 특정 태그의 첫번째 정보를 출력 
                    - find(속성명 = 속성값)
                        - 태그들 중 특정 속성이 특정한 값을 가진 태그의 첫번째 정보를 출력
                    - 함수의 결과 타입은 Tag 데이터로 되돌려준다.( Tag 데이터는 내장 함수가 사용이 가능 )
                - find_all()
                    - 문서 안에서 특정 태그의 모든 정보를 출력
                    - limit 매개변수 -> 정보들의 길이를 제한
                    - 함수의 결과는 ResultSet 타입으로 되돌려준다. ( Tag타입의 데이터들의 묶음 )
                    - ResultSet 타입에서는 find(), find_all() 함수 사용이 불가능 ( 각각의 원소는 타입이 Tag 이기때문에 함수 사용 가능 )
        - requests + bs4 두개의 라이브러리를 사용할때 한계점
            - 동적 타입의 웹 주소에서는 데이터를 가져오는데 문제점들이 발생 
            - ex)
                - 실시간으로 페이지가 변화하는 웹 서비스 ( http 통신은 요청, 응답을 받으면 서버와의 연결이 종료 )
                - 동적 페이지( 첫 요청시 페이지의 일부만 보여주는 경우 )

    - selenium 
        - 웹 페이지를 테스트 하기위한 라이브러리 
        - 웹 브라우져를 code를 이용해서 제어 하는 기능 
        - 크롤링에서는 동적 페이지의 html 문서를 가져오기 위해서 사용 
        1. 웹 브라우져를 selenium 안에 webdriver를 이용하여 실행 
        2. 특정 주소로 요청을 보낸다. (webdriver안에 get() 함수를 이용)
        3. 브라우져에서 열린 페이지의 태그를 찾는 함수들이 존재
            - find_element() -> BeautifulSoup 안에 find() 함수와 유사한 기능
            - find_elements() -> BeautifilSoup 안에 find_all() 함수와 유사한 기능
        4. 원하는 태그들을 찾아서 이벤트 발생 (특정 데이터를 입력, 마우스 클릭, 마우스 휠 이동)
        5. page_source 변수를 이용하여 브라우져의 html문서를 가져온다. 
        6. bs4 라이브러리를 이용하여 해당 페이지의 데이터를 추출
