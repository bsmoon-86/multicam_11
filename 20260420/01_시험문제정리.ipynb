```python
a = [1,2,3]
# list 안에 insert 함수 -> list에 데이터를 대입 (위치와 데이터를 인자값으로 사용)
a.insert(2, 4)
print(a)
```

```python
# 튜플 (1,2,3)에 원소를 추가하려면? 
# 튜플의 특징은 한번 생성을 하면 수정(추가, 삭제) 작업 불가 -> 새로 생성 
```

```python
def func(x, acc=[]):
    # 2개의 매개변수가 존재 
    # acc 빈 리스트는 가변의 데이터 -> 기본값을 그대로 사용하면 데이터들을 유지 (주소값)
    acc.append(x)
    return acc

print(func(1))
print(func(2))
print(func(3, []))
```

```python
a = [1,2,3]
b = a       # 얕은 복사 
b[0] = 100  # b의 첫번째 원소를 100 대입 -> [100, 2, 3] -> a도 변한다. 
print(a)
```
