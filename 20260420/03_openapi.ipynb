#### openapi
- api   
    - 웹 주소 
    - 웹 브라우져를 이용하여 특정 주소를 입력하였을때 데이터를 보내준다. 

```python
# openapi를 이용해서 데이터를 받아오기 위한 라이브러리 설치 
# !pip install requests
```

```python
import requests
from dotenv import load_dotenv
import os 
```

```python
load_dotenv()
```

```python
# 인증키 변수에 저장 ( dotenv를 이용해서 숨겨서 사용 )
service_key = os.getenv('service_key')
page_no = 1
num_of_rows = 100

url = 'https://apis.data.go.kr/B553069/esg/corp/getEsgStatus'

# 파라미터 지정 (dict형으로 key 값들은 openapi가 원하는 parameter의 key 값들을 지정)
params = {
    'serviceKey' : service_key, 
    'pageNo' : page_no, 
    'numOfRows' : num_of_rows
}

# 서버에게 요청을 보낸다 
# requests 안에 get( ) 함수를 이용
    # get()
        # 첫번째 인자 -> 주소
        # params 매개변수 -> 파라미터 값들을 dict 형태로 대입 

res = requests.get(url, params = params)
```

```python
res
```

```python
res.content
```

```python
type(res.content)
```

```python
# bytes 데이터 타입을 dict형태로 변환 라이브러리 호출 
import json
```

```python
# loads() 함수를 이용해서 bytes 데이터의 타입을 dict형으로 변환 
res_data = json.loads(res.content)
```

```python
type(res_data)
```

```python
from pprint import pprint
```

```python
pprint(res_data)
```

```python
# res_data에서 실제 데이터 부분을 추출하여 데이터프레임으로 생성 
import pandas as pd
```

```python
# res_data에서 items 안의 데이터들만 추출하여 데이터프레임으로 생성을 해보시오.
pprint(
    res_data['response']['body']['items']
)
```

```python
df = pd.DataFrame(res_data['response']['body']['items'])
df
```

```python
# openapi를 사용하기 위해 주소 
url = 'https://apis.data.go.kr/B010003/kamcoRlctRlst/pscd'

params = {
    'serviceKey' : os.getenv('service_key'), 
    'pageNo' : 1, 
    'numOfRows' : 100, 
    'resultType' : 'xml'
}

res = requests.get(url, params = params)
```

```python
res
```

```python
parmas2 = {
    'serviceKey' : os.getenv('service_key'), 
    'pageNo' : 1, 
    'numOfRows' : 100, 
    'resultType' : 'json'
}

res2 = requests.get(url, params = parmas2)
```

```python
res2
```

```python
res_data_json = json.loads(res2.content)
```

```python
pprint(res_data_json['body']['items']['item'])
```

```python
df2 = pd.DataFrame(res_data_json['body']['items']['item'])
df2.info()
```
